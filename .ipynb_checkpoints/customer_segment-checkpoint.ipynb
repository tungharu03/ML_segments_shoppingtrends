Tiến hành phân tích đặc trưng của khách hàng theo từng cụm để phân khúc khách hàng. Mục đích đưa ra những chiến lược khuyến mại hoặc thu hút khách hàng riêng biệt phù hợp cho từng nhóm



```python
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS
from IPython.display import Image
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('data\shopping_trends.csv')
df.sample(5)
```

```python
# Tạo từ điển ánh xạ cho "Frequency of Purchases"
frequency_mapping = {'Every 3 Months': 1, 'Annually': 2, 'Quarterly': 3, 'Monthly': 4, 
                     'Bi-Weekly': 5, 'Fortnightly': 6, 'Weekly': 7}
df['Frequency of Purchases'] = df['Frequency of Purchases'].map(frequency_mapping).astype(int)
```

# Tiến hành phân cụm tập dữ liệu

## 1. Sử dụng phương pháp khửu tay tìm số cụm tối ưu

```python
#Sử dụng 3 thuộc tính Age, Previous Purchases, Frequency of Purchases để phân cụm dữ liệu
data_cluster = df[['Age', 'Previous Purchases', 'Frequency of Purchases']]
data_cluster.head()
```

```python
#lấy biến độc lập:
X_cluster = data_cluster.values

#Hiển thị 10 mẫu đầu tiên:
X_cluster[:10,]
```

```python

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
#Thử số cụm (n_clusters) thay đổi từ 1 đến 20
for k in range(1, 21):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(X_cluster)  # X_cluster là dữ liệu của bạn cần gom cụm
    inertia.append(kmeans.inertia_)


# Marking the optimal number of clusters on the plot
optimal_clusters = 4  # Điều chỉnh dựa trên điểm khuỷu tay
plt.figure(figsize=(12,6))
plt.plot(range(1, 21), inertia, linewidth=2, color="red", marker="8")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS (Within Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal K")
plt.xticks(np.arange(1, 21, 1))
plt.axvline(x=optimal_clusters, color='blue', linestyle='--', linewidth=1)
plt.text(optimal_clusters + 0.1, max(inertia) * 0.9, 'No.of Optimal Clusters', color='blue', fontsize=10)
plt.show()

```

Từ biểu đồ trên theo phương pháp khửu tay ta thấy số lượng cụm n = 4 tốt nhất

## 2. Tiến hành phân cụm dữ liệu với số cụm đã xác định

```python
from sklearn.cluster import KMeans

#Thực hiện phân thành 4 cụm, các tham số khác sử dụng mặc định 
model_km = KMeans(n_clusters=4, random_state=0)

#Liệt kê các tham số của mô hình
model_km.get_params()
```

```python
#Thực hiện phân cụm
model_km.fit_predict(X_cluster)

#Tạo cột Cluster để cho biết mẫu thuộc cụm nào
data_cluster["Cluster"] = model_km.labels_
data_cluster
```

```python
#Thống kê số lượng khách hàng theo mỗi cụm:
sum_cluster = data_cluster.Cluster.value_counts()
print(sum_cluster)

#Trực quan hóa dữ liệu
plt.figure(figsize=(10,4))
plt.bar(sum_cluster.index, sum_cluster.values)
plt.title('THỐNG KÊ SỐ LƯỢNG KHÁCH HÀNG THEO TỪNG CỤM')
plt.grid()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection='3d')

# Trực quan hóa các cụm:
for i in range(4):
    ax.scatter(data_cluster["Age"][data_cluster["Cluster"] == i],
               data_cluster["Frequency of Purchases"][data_cluster["Cluster"] == i], 
               data_cluster["Previous Purchases"][data_cluster["Cluster"] == i], 
               s=100, label=f'Cluster {i}')

# Hiển thị tâm cụm:
centroids = model_km.cluster_centers_
for i in range(len(centroids)):
    ax.text(centroids[i, 0], centroids[i, 1], centroids[i, 2], 
            f'Centroid {i}', color='black', fontsize=12, ha='center')

ax.set_xlabel("Age")
ax.set_ylabel("Frequency of Purchases")
ax.set_zlabel("Previous Purchases")

plt.legend()
plt.show()

```

```python
#Vị trí tâm của các cụm
np.round(model_km.cluster_centers_,1)
```

```python
#Tổng bình phương khoảng cách từ các mẫu tới trung tâm cụm gần nhất.
model_km.inertia_
```
